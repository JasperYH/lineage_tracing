```python
def dist(a,b):
    if a=="0" and b=="0":
        return 1
    if a==b or a=="-" or b=="-":
        return 0
    if a=="0" or b=="0":
        return 2
    return 4
```

```python
import numpy as np
from Bio.Phylo.TreeConstruction import DistanceCalculator,DistanceMatrix,DistanceTreeConstructor
from Bio import Phylo

def build_nj_tree(df,out_file):
    n = len(df)
    m = len(df["seq"][0])
    dist_matrix = [[] for _ in range(n)]

    for i in range(n):
        for j in range(i):
            dist_matrix[i].append(sum([dist(df["seq"][i][k],df["seq"][j][k]) for k in range(m)]))
        dist_matrix[i].append(0)
    dist_matrix = dist_matrix
    dm = DistanceMatrix(df["label"].tolist(),dist_matrix)
    constructor = DistanceTreeConstructor()
    nj_tree = constructor.nj(dm)
    Phylo.write(nj_tree,out_file,"newick")
```

```python
from multiprocessing import Process
import pandas as pd
import os

samples = [i.split(".")[0] for i in os.listdir("../Data/Subchallenge2/Train/")]
processes = []
for sample in samples:
    print(sample)
    df = pd.read_table("../Data/Subchallenge2/Train/%s.txt"%sample,\
                   header=None,names=["label","seq"])
    p = Process(target=build_nj_tree, args=[df,"../Output/Subchallenge2/NJ_distmatrix/%s.nw"%sample])
    processes.append(p)
    p.start()

for p in processes:
    p.join()
```
